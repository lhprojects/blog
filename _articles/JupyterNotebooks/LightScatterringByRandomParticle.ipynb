# 光被粒子散射

```python
import LightScatteringByRandomParticle
%matplotlib inline
show_dia() # run appendix first
```

假设光被一团粒子散射。如果粒子的距离足够远（远远大于光的波长），并且位置是随机的，那么显然总的散射光强，是各个粒子散射光强之和。
但是如果各个粒子的平均距离和光的波长可比，那么计算总光强就需要考虑干涉的效应。

# 数学计算

入射光的波矢为$(k'_x,k'_y,k'_z)$，出射光的波矢为$(k''_x, k''_y, k''_z)$。定义
$$
(k_x,k_y,k_z) = (k''_x, k''_y, k''_z) - (k'_x,k'_y,k'_z)
$$


我们采用巨正则系综来来计算。
$p_i$是一个随机变量，其数值为$1$时表示$i$处存在一个粒子，其数值为$0$时表示$i$处不存在粒子，$\overline{p}_i$是$i$处存在一个粒子的概率。
$$
\sum_i \overline{p_i} = N
$$
单位密度的粒子数量为$P$，即


$$
\overline{p_i} = P(x_i,y_i,z_i) dx dy dz
$$ 
$$
\sum_i \overline{p}_i = \int P dx dy dz =  N
$$


我们计算散射光强，从下式开始
$$
A = \left|\sum _i\exp(i\phi_i)\right|^2
$$

其中
$$
\phi_i = z_i k_z + y_i k_y + x_i k_x
$$

继续计算得到


$$
A = \sum_{ii'} p_i p_{i'} e^{i(\phi_i -\phi_{i'})}
$$


利用


$$
\overline{ p_i^2 } =  \overline{p_i}\quad(二项式分布)
$$

$$
\overline{p_i p_{i'\neq i}} = \overline{p_i}\cdot\overline{p_{i'}}
$$

得到其期望值
$$
\overline{A} = N + \sum_{i\neq i'} \overline{p_i}\cdot\overline{p_{i'}} e^{i(\phi_i -\phi_{i'})}
$$


再利用
$$
\overline{p_i}
= dxdydz P
$$
得到

$$
\overline{A} = N + \Re\int\int dx'dy'dz' dxdydz P' P \exp(ik_z(z-z') + i k_y(y-y') + i k_x(x-x'))
$$

$$
\overline{A} = N + \int\int dx'dy'dz' dxdydz P' P \exp(ik_z z + i k_y y + i k_x x) \exp(-ik_z z - i k_y y - i k_x x)
$$

The (unitary, angular frequency) [fourier tranform](https://en.wikipedia.org/wiki/Fourier_transform) of $P$ is donated as $\widetilde{P}$
$$
\overline{A} = N + (2\pi)^{3} \widetilde{P}(k_x, k_y, k_z) \widetilde{P}(-k_x, -k_y, -k_z)
$$

利用 $N = (2\pi)^{3/2} \widetilde{P}(0,0,0)$

$$
\overline{A} = N \left(1 + N \frac{\widetilde{P}(k_x, k_y, k_z) \widetilde{P}(-k_x, -k_y, -k_z)}{\widetilde{P}^2(0, 0, 0)}\right)
$$

## 二维
当$k_x = 0$时，明显的，我们可以首先对$x$和$x'$积分。
可以将3维问题当作二维问题处理，并且二维的密度为$P(y,z)=\int dx P(x,y,z)$

$$
\bar{A} = N \left(1 + N \frac{\widetilde{P}(k_y, k_z) \widetilde{P}(-k_y, -k_z)}{\widetilde{P}^2(0, 0)}\right)
$$


## 干涉项对总体的影响
为了分析干涉项对总体的影响，我们将上式写为


$$
\bar{A} = N \left(1 + (2\pi)^{3/2}\frac{\widetilde{P}(k_x, k_y, k_z) \widetilde{P}(-k_x, -k_y, -k_z)}{\widetilde{P}(0, 0, 0)}\right)
$$

定义$I$

$$
\overline{A} = N \left(1 + (2\pi)^{3/2} I\right)
$$

考虑一种特殊情况，
设密度$P$可以写为
$$
P(x,y,z) = c^3 F(ax) F(ay) F(az)
$$

其中$c$的量纲为$L^{-1}$，
$a$的量纲为长度为$L^{-1}$，从而$F$本身无量纲。
而且并且假设$F$具有以下性质
$$
\widetilde{F}(k) = \widetilde{F}(-k) = G(k)
$$

$$
\widetilde{P}(\omega_1,\omega_2,\omega_3) = \prod_i c G(\omega_i)
$$

于是

$$
\widetilde{P}(\omega_1,\omega_2,\omega_3) = \prod_i c G(\omega_i/a)
$$


$$
I =\prod I_i = \prod_i c \frac{|G(\omega_i/a)|^2}{G(0)}
$$

当$a\rightarrow 0 $时，物质体积趋于无穷大。
我们预期$a\rightarrow 0$时候，对$\omega > 0，$我们有$I\rightarrow 0$。


### 密度分布：高斯

$$
 F(x) = {\displaystyle e^{-\alpha x^{2}}\,}	 (\alpha = a^2)
$$

$$
G(\omega) = {\displaystyle {\frac {1}{\sqrt {2\alpha }}}\cdot e^{-{\frac {\omega ^{2}}{4\alpha }}}}	
$$


$$
G(\omega)/G(0) = {\displaystyle e^{-{\frac {\omega ^{2}}{4\alpha }}}}	
$$
$$
I=  {\frac {c}{\sqrt {2\alpha }}} e^{-{\frac {\omega ^{2}}{2\alpha }}}
$$ 

$I$ 等于0的条件是

$$
{\frac {c}{\sqrt {2\alpha }}} e^{-{\frac {\omega ^{2}}{2\alpha }}} \rightarrow 0
$$

$$
\frac{L}{l_0} e^{-L^2/\lambda^2} \rightarrow 0
$$
其中$L$是物体的宏观尺度，$l_0$是粒子的平均距离，$\lambda = \lambda_0/\sin\theta$，$\lambda_0$为光的波长，$\theta$为散射角度。

我们可以看到，对于足够小的散射角度，干涉项总是不可忽略的。


### 密度分布：盒子
$$
F(x) = {\displaystyle \operatorname {rect} (ax)\,}		
$$

$$
G(\omega ) = {\displaystyle {\frac {1}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sinc} \left({\frac {\omega }{2\pi a}}\right)}	
$$

$$
G(\omega)/G(0) = {\displaystyle \operatorname {sinc} \left({\frac {\omega }{2\pi a}}\right)}	
= \frac{2\pi a}{\omega}\operatorname {sin} \left({\frac {\omega }{2\pi a}}\right)
$$

$$
I =  {\displaystyle {\frac {c}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sinc}^2 \left({\frac {\omega }{2\pi a}}\right)}	
= {\displaystyle \left(\frac {2\pi a}{\omega }\right)^2 {\frac {c}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sin}^2 \left({\frac {\omega }{2\pi a}}\right)}	
$$

$I$ 趋于0的条件是

$$
 \left(\frac {2\pi a}{\omega }\right)^2 {\frac {c}{\sqrt {2\pi a^{2}}}} \rightarrow 0
$$

$$
ac/\omega^2 \rightarrow 0
$$

$$
\lambda^2/(L l_0) \rightarrow 0
$$

其中$L$是物体的宏观尺度，$l_0$是粒子的平均距离，$\lambda = \lambda_0/ \sin\theta$，$\lambda_0$为光的波长，$\theta$为散射角度。

我们可以看到，对于足够小的散射角度，干涉项总是不可忽略的。

```python
%%file LightScatteringByRandomParticle.py

import matplotlib.pyplot as plt

def show_dia():
    np.random.seed(0)
    x = np.random.rand(5) - 0.5
    y = np.random.rand(5) - 0.5

    fig = plt.figure(figsize=(4,4))
    plt.scatter(x, y)
    plt.axis('off')
    plt.xlim((-0.2,0.75))
    plt.ylim((-0.2,0.75))
    for ax,ay in zip(x,y):
        plt.plot([-2,ax], [ay, ay], color="red")
        plt.plot([ax, 2], [ay, ay + (2-ax)], color="red")
    plt.text(0.75,0.25,r"$(k''_x,k''_y,k''_z)$")
    plt.text(-0.5,0.25,r"$(k'_x,k'_y,k'_z)$")
    #%mkdir  LightScatterringByRandomParticle
    #plt.savefig("LightScatterringByRandomParticle/lightscatterdia.png")
    plt.show()
    

    
    
    

import numpy as np
import numba
import matplotlib.pyplot as plt

# light come form -z 
# scatter to y
@numba.njit
def cal_phi(z, y, sintheta, costheta, k):
    phi = (z * (1 - costheta) - y*sintheta)*k
    return phi

@numba.njit
def MC_A(thetas, N, lamb, width_z, width_y, model):
    k = 2*np.pi/lamb
    
    coss = np.full_like(thetas, 0.)
    sins = np.full_like(thetas, 0.)
    
    sinthetas = np.sin(thetas)
    costhetas = np.cos(thetas)
    
    model_GAUSS = 0
    model_BOX = 1
    model_ = 0
    if model == "gauss":
        model_ = model_GAUSS
    elif model == "box":
        model_ = model_BOX
        
    for i in range(N):
        if model_ == model_GAUSS:
            z = np.random.normal()*width_z
            y = np.random.normal()*width_y
        elif model_ == model_BOX:
            z = (np.random.rand()-0.5)*width_z
            y = (np.random.rand()-0.5)*width_y
        
        for t, theta in enumerate(thetas): 
            phi = cal_phi(z, y, sinthetas[t], costhetas[t], k)
            #phi = 2*np.pi*np.random.rand()
            #print(z, y, phi)
            coss[t] += np.cos(phi)
            sins[t] += np.sin(phi)
            
    as_ = coss**2 + sins**2
            
    return as_/N

def cal_A(thetas, N, lamb, width_z, width_y, model):
    k = 2*np.pi/lamb
    
    As = np.full_like(thetas, 0.)
    
    ky = -np.sin(thetas)*k
    kz = (1-np.cos(thetas))*k
    if model == "gauss":
        def G_ratio(omega, width):
            alpha = 1/(2*width*width)
            return np.exp(-omega**2/(4*alpha))
    elif model == "box":
        def G_ratio(omega, width):
            a = 1/width
            x = omega/(2*np.pi*a)
            return np.sinc(x)
        
    return 1 + N*G_ratio(ky, width_y)*G_ratio(-ky, width_y)*G_ratio(kz, width_z)*G_ratio(-kz, width_z)
        
        
            

def plot(N, lamb, model="", 
         repeat=1, simulate=True,
         width_z=None, width_y=None, width=None,
         plot_range=1., plot_points=32, ax=None,
         plot_upper=False,
         title=None):
    
    N = int(N)
    thetas = np.linspace(-np.pi*plot_range, np.pi*plot_range,plot_points)    
    if width_z is None:
        width_z = width
    if width_y is None:
        width_y = width
    if width_z is None or width_y is None:
        assert False
        
    # simulate
    if simulate:
        Ass = np.empty(shape=(repeat, thetas.shape[0]))
        for r in range(repeat):
            Ass[r] = MC_A(thetas, N, lamb, width_z=width_z, width_y=width_y, model=model)
        As = Ass.mean(axis=0)
    
    Bs = cal_A(thetas, N, lamb, width_z=width_z, width_y=width_y, model=model)
    
    
    # begin plot
    show=False
    if ax is None:
        plt.figure(figsize=(4,4))
        ax = plt.gca()
        
    if simulate:
        for r in range(repeat):
            #ax.plot(thetas, Ass[r,:], label="numerical")
            pass
        ax.plot(thetas, As, label="numerical")
            
        if repeat >= 2 and plot_upper:
            VarAs = Ass.std(axis=0)
            ax.plot(thetas, As + VarAs, label="numerical + 1$\sigma$", linestyle="--")
            
        
        
    ax.set_xlabel(r"$\theta$")
    ax.plot(thetas, Bs, label="theory")
    ax.set_yscale("log")
    ax.set_title(title)
    ax.legend()
    
    if show:
        plt.show()
```
