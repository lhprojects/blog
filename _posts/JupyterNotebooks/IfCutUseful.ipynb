If cut is useful:
$$
S/\sqrt{S+B}
<S\epsilon_S/\sqrt{S\epsilon_S+B\epsilon_B}
$$

$$
\epsilon_S^2(S+B)>(S\epsilon_S+B\epsilon_B)
$$

$$
\epsilon_S^2 - S/(S+B) \epsilon_S  - B\epsilon_B/(S+B) > 0
$$


```python
import numpy as np
import matplotlib.pyplot as  plt
def sovle_epsilonS(S,B,epsilonB):
    a = 1
    b = -S/(S+B)
    c = -epsilonB*B/(S+B)
    delta = b*b-4*a*c
    return (-b+np.sqrt(delta))/(2*a)
```

```python
EpsilonBackgrounds = 0.5
MinRatioOfB2S = 0.5
MaxRatioOfB2S = 100

plt.figure(figsize=(5,5))
plt.plot(np.linspace(MinRatioOfB2S,MaxRatioOfB2S),
         sovle_epsilonS(1, np.linspace(MinRatioOfB2S,
        MaxRatioOfB2S), EpsilonBackgrounds))
plt.xlabel("B/S", fontsize=20)
plt.ylabel("$\epsilon_S$ required ($\epsilon_B=%.2f$)"%(EpsilonBackgrounds), fontsize=20)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
RatioOfB2S = 10

plt.figure(figsize=(5,5))
plt.plot(np.linspace(0, 1), sovle_epsilonS(1, RatioOfB2S, np.linspace(0, 1)))
plt.xlabel("$\epsilon_B$", fontsize=20)
plt.ylabel("$\epsilon_S$ required ($B/S=%.2f$)"%RatioOfB2S, fontsize=20)
plt.grid()
plt.tight_layout()
plt.show()
```
