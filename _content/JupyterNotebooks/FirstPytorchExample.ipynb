# MNIST数据集

```python
import torch
import matplotlib.pyplot as plt

from torchvision import datasets, transforms

transform=transforms.Compose([
    transforms.ToTensor(),
    ])

train_dataset = datasets.MNIST('../data', train=True, download=True,
                       transform=transform)
test_dataset = datasets.MNIST('../data', train=False,
                       transform=transform)

print(train_dataset)
print(test_dataset)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32)

print("number of training batchs:", len(train_loader))
print("number of testing batchs:",len(test_loader))
for sample in train_loader:
    print("input size (batch_size, channels, width, height):", tuple(sample[0].size()))
    print("output size (batch_siz,e):", tuple(sample[1].size()))
    break
```

# first batch

```python
for batch in train_loader:
    fig, axes = plt.subplots(4, 8, figsize=(8,4))
    for i in range(4):
        for j in range(8):
            input = batch[0]
            output = batch[1]
            n = 4*i + j
            axes[i][j].imshow(input[n,0])
            axes[i][j].set_title(str(int(output[n])))
            axes[i][j].set_xticks([])
            axes[i][j].set_yticks([])
    plt.tight_layout()
    plt.show()
    break
```
