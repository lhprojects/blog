# 高维二次样条回归

## 一元二次（自然）样条函数
 
 
 二次样条函数
 
* 分段二次函数，分段点是$\xi_{0\leq i<K-1}$，一共有K个分段点
* 连续函数
* 导数连续
 
自然样条函数还应该满足
* 在区间$[\xi_{min}=x_0,\xi_{max} = x_{K-1}]$外，二次以及以上导数为0。
 
二次自然样条可以表示基矢函数的叠加
$$
f(\mathbf{x}) = \sum_{0\leq i<N} \beta_i h_i(\mathbf{x})
$$

总共有$N=K+1$个基矢，每个基矢也应该满足自然二次样条函数的条件。




## B样条


## 高维样条函数
设在D维空间中，一个数据点 $\mathbf{x} = (x_0, ... , x_D)^T$，高维样条函数表示为一维样条函数的乘积，编号为$(i,j,...,k)$的样条函数为
$$
h_{ij..k}(\mathbf{x}) = h^{(0)}_i(x_0) h^{(1)}_j(x_1)  .. h^{(D-1)}_k(x_{D-1})
$$
设第$k$维空间有$N^{(k)}$个样条函数，则总样条函数为
$$
N = \prod_{k<D} N^{(k)}
$$


## Loss函数

$$
L = \sum_{j < M}(y_j - f(\mathbf{x}_j))^2 
+ \alpha M \sum_{0<i<N} \beta_i^2 
$$
其中$M$为样本数量。
其中$N$为样条基矢数量。
注意L2正则项不包含对第一个样条（即常数项截距项）的惩罚项。

$f(\mathbf{x})$为样条函数
$$
f(\mathbf{x}) = \sum_{0\leq i<N} \beta_i h_i(\mathbf{x})
= \sum_{0\leq i<N} \beta_i \prod_{j<D} h^{(j)}_{m_i}(x_j)
$$
其中$D$为空间维度。$h_i(\mathbf{x}) = h_{m_i n_i...o_i}(\mathbf{x})$。$i$是样条函数的紧凑编号，$m_i n_i...o_i$是样条函数的展开编号。一共有$N$个样条
$$
N = \prod_{k<D} N^{(k)}
$$



## 实现

```python
#%%file quadratic_spline_regression.py

import numpy as np
import scipy
import warnings

__all__ = ["QuadraticSplineRegressor",\
           "QuadraticSplineKernel",\
           "QuadraticSplineKernel_Ridge",\
           "QuadraticSplineKernel_GP"]

def _cal_N_bases(bases):    
        N_bases = 1
        for bases_1D in bases:
             N_bases *= bases_1D.N_bases 
        return  N_bases

    
def _cal_H(X, bases, N_bases):
    '''
    return N_data x N_bases array
    '''
        
    splines = bases
    N_dims = len(splines)
    N_splines = N_bases

    Ndata = X.shape[0]
    assert Ndata > 0

    #print("N_splines", N_splines)
    #print("Ndata", Ndata)

    H = np.ones((Ndata, N_splines))

    dim_in = 1
    for d in range(N_dims-1,-1,-1):

        x = X[:, d]
        splines_1D = splines[d]
        N_splines_1D = splines_1D.N_bases                        
        # [Ndata, N_splines_1D]
        h = splines_1D.get_values(x)

        # H[i][j][k] = h1[i]h2[j]h3[k]
        # H[i*JK + j*K + k] = h1[i]h2[j]h3[k]
        # H[i*JK+ j*stride + offset] = ... h2[j]...            
        dim_out =  N_splines//N_splines_1D//dim_in
        #print("d", d)
        #print("N_splines", N_splines)
        #print("dim_in", dim_in)
        #print("N_splines_1D", N_splines_1D)
        H = np.reshape(H, (Ndata, dim_out, N_splines_1D, dim_in))

        # don't understand?
        # take a numpy course
        H[:,:,:,:] *= h[:,np.newaxis,:,np.newaxis]

        dim_in *= N_splines_1D


    H = H.reshape(Ndata, N_splines)
    return H


class TensorBasesKernel:
    
    def __init__(self):
        pass
    
    def diag(self, X):
        if not hasattr(self, "bases") or self.bases is None:
            self.initialize_bases()
            
        H = _cal_H(X, self.bases, self.N_bases)
        return np.square(H).sum(axis=1)
    
    def is_stationary(self):
        return False
    
    def __call__(self, X, Y=None, eval_gradient=False):
        '''
        X is a matrix of shape of N_data x N_feature        
        Y is a matrix of shape of N_data x N_feature        
        return phi(X)^T \cdot phi(Y)
        it's a matrix of shape of N_data x N_data
        '''
        if not hasattr(self, "bases") or self.bases is None:
            self.initialize_bases()
        
        H1 = _cal_H(X, self.bases, self.N_bases)        
        # we don't l2 regularization on bias
        H1[:,0] *= 1.0
        
        if Y is not None:
            H2 = _cal_H(Y, self.bases, self.N_bases)
            H2[:,0] *= 1.0
        else:
            Y = X
            H2 = H1                    
        
        kernel = H1.dot(H2.T)
        
        if eval_gradient:
            # we don't parmaeters, return an empty array for gradient
            return kernel, np.zeros((kernel.shape[0],kernel.shape[1],0)) 
        else:
            return kernel
        
        
        
class TensorBasesPredictor:
    
    def __init__(self, bases, N_bases, beta, intercept):
        self.bases = bases
        self.N_bases = N_bases
        self.beta = beta
        self.intercept = intercept
            
    def predict(self, X):
        
        if type(X) is not np.ndarray:
            # let's have a float copy
            X = np.array(X, dtype=np.float_)
            
        if X.ndim == 1:
            X = X.reshape(-1, 1)
            
        elif X.ndim != 2:
            raise Exception("Data format incorrect")
            
        # no data
        if X.shape[0] == 0:
            raise Exception("No data to predict!\n")
            
        N_dims = len(self.bases)       
        if N_dims != X.shape[1]:
            raise Exception("dimension not match data's! N_dims = %d != X shpae[1]  = %d"%(N_dims, X.shape[1]))
                    
        # real work here
        #   y = sum_i H_i( x) beta_i + b
        
        H = _cal_H(X, self.bases, self.N_bases)
        yhat = H.dot(self.beta) + self.intercept
        return yhat


    
class TensorBasesRegressor:
    
    def __init__(self, force_lstsq=True):
        self.force_lstsq = force_lstsq
    
    def initialize(self):
        
        self.alpha_ = self.alpha
        if self.alpha is None:
            self.alpha_ = 1E-6
                    
        self.initialize_bases()
                    
    def update_X_and_y(self, X, y):
    
        alpha = self.alpha_
        N_dims = self.N_dims
        N_bases = self.N_bases
        
        if type(X) is not np.ndarray:            
            X = np.array(X, dtype=np.float_)
            
        if type(y) is not np.ndarray:            
            y = np.array(y, dtype=np.float_)
            
        if X.ndim == 1:
            X = X.reshape(-1, 1)
            
        elif X.ndim != 2:
            raise Exception("data format incorrect")
            
        # no data
        if X.shape[0] == 0:
            raise Exception("no data to fit!\n")
            
        if N_dims != X.shape[1]:
            raise Exception("dimension not match data's! N_dims = %d != X shpae[1]  = %d"%(N_dims, X.shape[1]))
        
                    
        N_data = np.shape(X)[0]
                
        H = _cal_H(X, self.splines, N_bases)
               
        # center data
        H_mean = np.mean(H, axis=0)
        H -= H_mean
        
        # center data
        y_mean = np.mean(y, axis=0)
        y_ = y - y_mean

        # now we minimize
        #   sum_j (y_j - sum_i H_i(X_j) beta_i)^2 + lambda sum_i beta_i^2
        # the solution is
        #  beta = (H^T H  + lambda I)^-1 H^T y

        HTH = H.T.dot(H)
        
        # fill the diagal
        lamb = alpha * N_data
        HTH.flat[0::N_bases+1] += lamb
                    
        HTy = H.T.dot(y)
        
        if self.force_lstsq:
            beta,_,_,_ = scipy.linalg.lstsq(HTH, HTy)
        else:
            try:
                beta = scipy.linalg.solve(HTH, HTy, sym_pos=True)
            except scipy.linalg.LinAlgError:
                warnings.warn("failed to solve linear equation! We will try least squre method!")
                beta,_,_,_ = scipy.linalg.lstsq(HTH, HTy)
                
                                
        self.beta = beta
        #self.intercept = 0
        self.intercept = y_mean - H_mean.dot(beta) 
            

            
    def fit(self, X, y):
        self.initialize()        
        self.update_X_and_y(X, y)
                
        
    def predict(self, X):
        
        if type(X) is not np.ndarray:
            # let's have a float copy
            X = np.array(X, dtype=np.float_)
            
        if X.ndim == 1:
            X = X.reshape(-1, 1)
        elif X.ndim != 2:
            raise Exception("Data format incorrect")
            
        # no data
        if X.shape[0] == 0:
            raise Exception("No data to predict!\n")
            
        N_dims = self.N_dims        
        if N_dims != X.shape[1]:
            raise Exception("dimension not match data's! N_dims = %d != X shpae[1]  = %d"%(N_dims, X.shape[1]))
                    
        # real work here
        #   y = sum_i H_i( x) beta_i + b
        
        H = _cal_H(X, self.splines, self.N_bases)
        yhat = H.dot(self.beta) + self.intercept
        return yhat
    
    
    def score(self, X, y):
        yhat = self.predict(X)        
        return 1 - np.sum( np.square(y-yhat) ) / np.sum( np.square(y-y.mean()) )


    
    def get_predictor(self, output = None):        
        if self.beta.ndim == 1:
            assert output is None
            return TensorBasesPredictor(self.splines, self.N_bases, self.beta, self.intercept)
        else:
            assert self.beta.ndim == 2
            if output is not None:
                return TensorBasesPredictor(self.splines, self.N_bases, self.beta[output], self.intercept)
            else:
                return TensorBasesPredictor(self.splines, self.N_bases, self.beta, self.intercept)
                
    

    
    



    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
class BSpline1D:
    
    def __init__(self, knots, degree):
        assert np.ndim(knots) == 1
        knots = np.array(knots, dtype=np.float_)
        knots = np.sort(knots)        
        self.N_bases = degree + len(knots)-1
        self.knots = knots
        self.degree = degree
        
        
    def get_values(self, X):
        '''
            return array of form N_data * N_bases
        '''
        
        if type(X) is not np.ndarray:
            X = np.array(X)
        
        knots = self.knots
        N_bases = self.N_bases
        N_data = np.shape(X)[0]                              
        degree = self.degree
        
        # at least 2 knots
        assert len(knots) >= 2
        
        xi_max = knots[-1]
        xi_min = knots[0]
        
        
        if True:
            knots_ = np.concatenate(( np.full(degree, knots[0]), knots, np.full(degree, knots[-1]) ))
        elif False:
            knots_ = np.concatenate(( knots[0] - np.arange(degree, 0, -1)*(knots[1] - knots[0]),\
                                     knots,\
                                     knots[-1] + np.arange(1, degree+1)*(knots[-1] - knots[-2]) ))
        elif False:
            knots_ = np.concatenate(( knots[0] - np.arange(degree, 0, -1)*1E6,\
                                     knots,\
                                     knots[-1] + np.arange(1, degree+1)*1E6 ))
        elif False:
            if degree == 0:
                knots_ = knots
            elif degree == 1:
                knots_ = np.concatenate((knots[0:1],\
                                         knots,\
                                         knots[-1:]))
            elif degree >= 2:
                knots_ = np.concatenate(( knots[0] - np.arange(degree-1, 0, -1)*1E6,\
                                         knots[0:1],\
                                         knots,\
                                         knots[-1:],\
                                         knots[-1] + np.arange(1, degree)*1E6 ))
        elif False:
            if degree == 0:
                knots_ = knots
            elif degree == 1:
                knots_ = np.concatenate((knots[0:1] -(knots[1] - knots[0]),\
                                         knots,\
                                         knots[-1:] + (knots[-1] - knots[-2])))
            elif degree >= 2:
                knots_ = np.concatenate(( knots[0] - np.arange(degree-1, 0, -1)*1E6,\
                                         knots[0:1] - (knots[1] - knots[0]),\
                                         knots,\
                                         knots[-1:] + (knots[-1] - knots[-2]),\
                                         knots[-1] + np.arange(1, degree)*1E6 ))
                
                
                
        
        B0 = np.zeros(( N_data, len(knots_)-1 ))        
        for i in range(B0.shape[1]):
            B0[:,i] = np.where(X < knots_[i], 0, np.where(X >= knots_[i+1], 0, 1))
        
        for k in range(1, degree+1):            
            
            B1_N_bases = B0.shape[1] - 1
            B1 = np.zeros((N_data, B1_N_bases))
            
            for i in range(B1_N_bases): 
                if knots_[i+k] == knots_[i] and knots_[i+k+1]== knots_[i+1]:
                    B1[:, i] = 0
                elif knots_[i+k] == knots_[i]:      
                    B1[:,i] = (knots_[i+k+1]-X)/(knots_[i+k+1]-knots_[i+1]) * B0[:,i+1]                
                elif knots_[i+k+1]== knots_[i+1]:
                    B1[:,i] = (X-knots_[i])/(knots_[i+k]-knots_[i]) * B0[:,i]           
                else:
                    B1[:,i] = (X-knots_[i])/(knots_[i+k]-knots_[i]) * B0[:,i] + \
                         (knots_[i+k+1]-X)/(knots_[i+k+1]-knots_[i+1]) * B0[:,i+1]
                    
            B0 = B1            
            
    
            
        assert B0.shape == (N_data, N_bases)
        return B0

    
class BSplineRegressor(TensorBasesRegressor):
    
    def __init__(self, knots=None, alpha=None, degree=3, force_lstsq=True):
        self.knots = knots
        self.alpha = alpha
        self.degree = degree
        super().__init__(force_lstsq)
        
        
    def get_params(self, deep=False):
        return {
            "knots":self.knots,
            "alpha": self.alpha,
            "degree": self.degree,
        }
    
    def set_params(self, **kargs):
        for key in kargs:
            if key == "knots":                
                self.N_dims = None
                self.N_bases = None
                self.splines = None
                self.knots_ = None
                
                self.H = None
                
                self.Omega = None
                self.spline_sub_index = None
                
                self.knots = kargs[key]
            elif key =="degree":
                self.degree = kargs[key]
            elif key == "alpha":
                self.alpha = kargs[key]
            else:
                raise(Exception("Unkown arg %s"%key))
        return self
    
    def __repr__(self):
        return 'QuadraticSplineRegressor(knots[%d]=%s,alpha=%.1E)'%(len(self.knots), str(self.knots), self.alpha)
        
                
    def initialize_bases(self):
                
        if not hasattr(self, "splines") or self.splines is None:
            knots = self.knots        
            assert knots is not None

            all_is_scalar=True
            all_is_list=True
            for item in knots:
                all_is_scalar = all_is_scalar and np.isscalar(item)
                all_is_list = all_is_list and np.ndim(item) == 1

            if all_is_scalar:
                if len(knots) < 1:
                    raise Exception("At lease one knot")
                knots = np.array(knots).reshape(1,-1)
            elif all_is_list:
                pass
            else:
                raise Exception('''1d array or 2d array knots needed, format: [xi0, xi1, ...] or \
                    [[xi00, xi01, ...],\
                    [xi10, xi11, ..],\
                    ...]''')

            self.splines = [BSpline1D(knots_1D, degree=self.degree) for knots_1D in knots]                    
            self.N_bases = _cal_N_bases(self.splines)
            self.N_dims = np.shape(knots)[0]
            self.knots_ = knots

    
    
    
    
    
    
    
class BSplineKernel(TensorBasesKernel):
    
    def __init__(self, knots, degree=3):
        self.knots = knots
        self.degree = degree
        
    def get_params(self, deep=False):
        return { "knots": self.knots, 
                "degree": self.degree,
               }
    
    def set_params(self, **kargs):
        for key in kargs:
            if key == "degree":
                self.degree = kargs[key]   
                self.bases = None
            elif key == "knots":
                knots = kargs[key]
                self.bases = None
            else:
                raise Exception("unkown parmaeter %s"%key)
        
        return self
    
    def initialize_bases(self):
        self.bases = [BSpline1D(knots_1D, self.degree) for knots_1D in self.knots]
        self.N_bases = _cal_N_bases(self.bases)
    
    
    
class BSplineKernel_Ridge(BSplineKernel):
    
    def __call__(self, x1, x2):        
        return super().__call__(np.array([x1]), np.array([x2]))[0]
  
    
try:
    from sklearn.gaussian_process.kernels import Kernel
    class BSplineKernel_GP(BSplineKernel, Kernel):
        pass
except:
    pass
```

## 基矢函数

```python
def plot_bases(knots_, degree):
    
    Nknots = len(knots_)
    fig, axes = plt.subplots(1, Nknots, figsize=(4*Nknots,4))    
    
    for ax, knots in zip(axes, knots_):
        
        plt.sca(ax)
        xlim=[-3, 5]
        ylim=[-1, 2]

        X_plt = np.linspace(xlim[0], xlim[1], 100)



        reg = BSplineRegressor(knots=knots, degree=degree)
        reg.initialize_bases()
        reg.beta = np.zeros(reg.N_bases)

        for knot in reg.knots_:
            plt.plot([knot, knot], [-4,4], linestyle="--", color="blue")
            
        for i in range(reg.N_bases):
            reg.beta[:]=0
            reg.beta[i]=1
            reg.intercept=0
            plt.plot(X_plt, reg.predict(X_plt), label="basis %d"%i)        
    
        plt.xlabel("$x$", fontsize="large")
        plt.ylabel("$y$", fontsize="large")
        plt.xlim(xlim)
        plt.ylim(ylim)
        plt.legend()
        plt.grid()
        
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def fit_plot(X_train, y_train, knots, xlim, ax, ylim=None, alpha=None, label="", alphac=None, add_label=True, degree=3):    
    X_plt = np.linspace(xlim[0], xlim[1], 100)
    
    reg = BSplineRegressor(knots=knots, alpha=alpha, degree=degree)
    reg.fit(X_train, y_train)
    
    if add_label:
        ax.scatter(X_train, y_train, label="data", color="blue", s=20)
    else:
        ax.scatter(X_train, y_train, color="blue", s=20)
                
    ax.plot(X_plt, reg.predict(X_plt), label=label)
    
    for i, knot in enumerate(knots):
        if i == 0 and add_label:
            ax.plot([knot, knot],[-4,4], linestyle="--", color="blue", label="knot")
        else:
            ax.plot([knot, knot],[-4,4], linestyle="--", color="blue")

    ax.set_xlabel("$x$", fontsize="large")
    ax.set_ylabel("$y$", fontsize="large")
    ax.set_ylim(ylim)
    ax.set_xlim(xlim)
    ax.legend()
    
```

## L2正则化参数 $\alpha$ 
$\alpha$ 倾向于比较小的样条函数系数，但不包含常数项。

```python
def f(x):
    x = np.array(x)
    return -1/2*x*x +x + 1
x = [0.5, 1.5, 1.6, 2.5]
y = f(x)
y_with_noise1 = y + [0,0, 0.05,0]
y_with_noise2 = y + [0,0,-0.05,0]

fig = plt.figure(figsize=(16,4))

for i in range(4):
    ax = fig.add_subplot(1,4,1+i)
    if i == 0:
        alpha = 0
        ax.set_title(r"$\alpha$=0, overfit")
    elif i == 1:
        alpha = 1E-3
        ax.set_title(r"$\alpha$=%.1E, good"%alpha)
    elif i == 2:
        alpha = 0.1
        ax.set_title(r"$\alpha$=%.1E, underfit"%alpha)
    elif i == 3:
        alpha = 1000
        ax.set_title(r"$\alpha$=%.1E, shinkage limit"%alpha)
        
    fit_plot(X_train = x, y_train =y, xlim=(-1,4),ylim=(-2,4), knots=[0, 1.5, 3],\
             alpha=alpha, ax=ax, label="regress data wo/ noise", add_label=False, degree=2)
    fit_plot(X_train = x, y_train =y_with_noise1, xlim=(-1,4),ylim=(-2,4), knots=[0, 1.5,3],\
             alpha=alpha,  ax=ax, label="regress data w/ noise", add_label=False, degree=2)
    fit_plot(X_train = x, y_train =y_with_noise2, xlim=(-1,4),ylim=(-2,4), knots=[0, 1.5, 3],\
             alpha=alpha,  ax=ax, label="regress data w/ noise", add_label=True, degree=2)
    ax.plot(np.linspace(-1,4), f(np.linspace(-1,4)), label="truth")
    ax.legend()

plt.tight_layout()
plt.show()
```

## 高密度knots


如果knots密度很高，且正则化参数很弱，显然曲线自由度会很高，曲线会穿过所有数据点。

```python
def f(X):
    return np.exp(X*np.cos(np.pi*X))
    
X_plt = np.linspace(0,1,1000)
X_train = np.linspace(0,1,5)

plt.figure(figsize=(8,4))
plt.plot(X_plt, f(X_plt), label=r"truth $y(x)$")
plt.scatter(X_train, f(X_train), label="traning")

qsr1=BSplineRegressor(knots=np.linspace(0, 1.0, 1000), alpha=0.01)
    
    
qsr1.fit(X_train, f(X_train))

plt.plot(X_plt, qsr1.predict(X_plt), label=r"$\hat{y_1}(x)$: L2")
plt.ylabel(r"$y$")
plt.xlabel("$x$")
plt.ylim((0,2.7))
handlers, labels = plt.gca().get_legend_handles_labels()

plt.sca(plt.gca().twinx())
plt.ylabel(r"$\Delta y$")
handlers2, labels2 = plt.gca().get_legend_handles_labels()

handlers += handlers2
labels += labels2
plt.legend(handlers, labels,fontsize="large")
plt.show()
```

## $\exp(x\cos(3\pi x))$

数据并没有随机误差，我们取足够多的knots，此时回归退化成了插值。（注意预测与函数真实值在数据点附近差值为0）

* 采用L2或者曲率正则化的样条回归与函数真实值的最大误差为0.008

* 没有正则化高斯核回归方法的最大误差0.027 (see https://en.wikipedia.org/wiki/Radial_basis_function_interpolation)。

* 采用高斯核的ridge regression性能和样条回归差不多。最大误差为0.006。

```python
from sklearn.kernel_ridge import KernelRidge
import sklearn.metrics.pairwise as pairwise

def exp_x_cos_3pi_x(scale):

    X_plt = np.linspace(0,1,1000)
    X_train = np.linspace(0,1,15)

    X_train = X_train*scale
    X_plt = X_plt*scale
    knots=np.linspace(0, 1, 10)*scale


    def f(X):
        return np.exp(X/scale*np.cos(3*np.pi*X/scale))

    plt.figure(figsize=(8,4))
    plt.plot(X_plt, f(X_plt), label=r"$y(x)$: truth")
    plt.scatter(X_train, f(X_train), label="traning")

    qsr=BSplineRegressor(knots=knots, alpha=1E-5, degree=3)
    # sklearn.metrics.pairwise.rbf_kernel
    # gamma definition: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.rbf_kernel.html
    # set alpha=0 to reproduce the result in wikipeida
    kr = KernelRidge(alpha=1E-7, kernel="rbf", gamma=3.4*3.4)

    for k in qsr.knots:
        plt.plot([k,k], [0,3], linestyle="--", color="green")

    qsr.fit(X_train, f(X_train))
    kr.fit(X_train.reshape(-1,1), f(X_train))

    plt.plot(X_plt, qsr.predict(X_plt), label=r"$\hat{y}_1(x)$: quadratic spline")
    plt.plot(X_plt, kr.predict(X_plt.reshape(-1,1)), label=r"$\hat{y}_2(x)$: gaussian kernel ridge")
    plt.ylabel(r"$y$")
    plt.xlabel("$x$")
    plt.ylim((0,2.7))
    plt.xlim((0,1*scale))
    handlers, labels = plt.gca().get_legend_handles_labels()

    plt.sca(plt.gca().twinx())
    plt.plot(X_plt, qsr.predict(X_plt) - f(X_plt),  label="$\hat{y}_1(x) - y(x)$", linestyle="--")
    plt.plot(X_plt,  kr.predict(X_plt.reshape(-1,1)) - f(X_plt), label="$\hat{y}_2(x) - y(x)$", linestyle="--")
    plt.ylabel(r"$\Delta y$")
    handlers2, labels2 = plt.gca().get_legend_handles_labels()

    handlers += handlers2
    labels += labels2
    plt.subplots_adjust(right=0.6)
    plt.legend(handlers, labels, fontsize="large", loc=(1.3,0.2))
    plt.show()

exp_x_cos_3pi_x(1.0)
```

## 尺度不敏感

如果$\alpha_c = 0$, 拟合对尺度不敏感。

我们已经在算法内部通过`knots`对`X`进行了缩放。
所以，对于X的某一个维度`i`，对`X[:,i]`和`knots[i]` (`knots`是二维数组，`knots[i]`表示第`i`维的所有`knots`)同时缩放，不会影响最终的结果。
所以你只需要对每一个维度设置合适的knots就可以了。这对于调节超参数来说更方便一点。

但是对于使用高斯核心的回归算法。你需要对每一个维度进行合适的缩放。如果你要拟合一维函数，也可以不缩放数据，只调节`gamma`，这对于调节超参数来说更方便一点。

# 2D fit

```python
from mpl_toolkits.mplot3d import Axes3D
def fit_plot3d_reg(X_train, y_train, knots, reg, X_test = None, y_test = None,\
                   xlim=(-1,1), ylim=(-1,1), zlim=(-1,1), ax=None, sz=20):
    y_train =np.array(y_train)
    X_train = np.array(X_train)
    
    reg.fit(X_train, y_train)
    
    
    
    x_plt = np.linspace(xlim[0], xlim[1], 100)
    y_plt = np.linspace(ylim[0], ylim[1], 100)
    xx, yy = np.meshgrid(x_plt, y_plt)
    yhat_train = reg.predict(X_train)
    up = y_train > yhat_train
    down = y_train <= yhat_train
    
    if np.any(up):
        ax.scatter(X_train[:,0][up], X_train[:,1][up], y_train[up], s=sz, color="red")
    if np.any(down):
        ax.scatter(X_train[:,0][down], X_train[:,1][down], y_train[down], s=sz, color="green")
    
    from matplotlib import cm
    ax.plot_surface(xx, yy, reg.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(100,100), alpha=0.3, cmap=cm.coolwarm)
    
    if knots is not None:
        for k1 in knots[0]:
            ax.plot([k1, k1],[-1,1],[0,0], linestyle="--", color="blue")
        for k2 in knots[1]:
            ax.plot([-1, 1],[k2, k2],[0,0], linestyle="--", color="blue")
        
    ax.set_xlabel("$x_1$")
    ax.set_ylabel("$x_2$")
    ax.set_zlabel("$y$")
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_zlim(zlim)
    err = y_train-reg.predict(X_train)
    mse = np.sqrt(np.square(err).mean())
    if X_test is not None and X_test is not None:
        err = y_test-reg.predict(X_test)
        mse2 = np.sqrt(np.square(err).mean())
        maxe = np.max(np.abs(err))
        ax.set_title("RMSE(Train) = %.1E,\nRMSE(Test) = %.1E,\nMAX_ERR(Test) = %.1E"%(mse, mse2, maxe))
    else:
        maxe = np.max(err)
        ax.set_title("RMSE(Train) = %.1E,\nMAX_ERR(Train) = %.1E"%(mse, maxe))
    
    return reg
```

## 基矢函数：以3x3分段网格为例

```python
def fit_plot3d_i(knots, i, xlim=(-1,1), ylim=(-1,1), zlim=(-1,1), ax=None, sz=20):    
    
    x_plt = np.linspace(xlim[0], xlim[1], 100)
    y_plt = np.linspace(ylim[0], ylim[1], 100)
    xx, yy = np.meshgrid(x_plt, y_plt)
    reg = BSplineRegressor(knots=knots)
    #beta = reg.beta[i]
    reg.initialize()
    reg.beta = np.zeros(reg.N_bases)
    reg.beta[i] = 1
    reg.intercept = 0
    #print(reg.beta)
    #print(reg.bias)
    
    from matplotlib import cm
    ax.plot_surface(xx, yy, reg.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(100,100), alpha=0.2, cmap=cm.coolwarm)
    
    for k1 in knots[0]:
        ax.plot([k1, k1],[-1,1],[0,0], linestyle="--", color="blue")
    for k2 in knots[1]:
        ax.plot([-1, 1],[k2, k2],[0,0], linestyle="--", color="blue")
        
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_zlabel("$z$")
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    #ax.set_zlim(zlim)
    ax.set_title("Spline %d of %d"%(i+1, reg.N_bases))
    
    return reg


fig=plt.figure(figsize=(12,12))

for i in range(16):

    ax1 = fig.add_subplot(4,4,i+1, projection="3d")
    fit_plot3d_i(i=i, zlim=(-2,2), knots=[np.linspace(-1,1,3)]*2, ax=ax1, sz=10)

plt.tight_layout()
plt.show()
```

```python
def f21(X):
    #return X[:,0]*X[:,0] + 1/2*X[:,1]**3 + 0*np.random.randn(X.shape[0])
    return X[:,0]**4 + 1/(1+0.3*X[:,1]) + 0*np.random.randn(X.shape[0])


Ndata=100*100
N=2
N_grid=10

X_rnd=2*np.random.rand(Ndata,N)-1

X_grid=np.zeros((N_grid*N_grid, N))
c=0

x1=np.linspace(-1,1,N_grid)
x2=np.linspace(-1,1,N_grid)

for i in range(N_grid):
    for j in range(N_grid):
        X_grid[c,0] = x1[i]
        X_grid[c,1] = x1[j]
        c += 1

X_train = X_grid     
y_train = f21(X_train)
X_test = X_rnd
y_test = f21(X_test)


fig=plt.figure(figsize=(16,4))
for i in range(4):
    ax1 = fig.add_subplot(1,4,1+i, projection="3d")
    reg = fit_plot3d_qsr(X_train = X_train, y_train = y_train, X_test=X_test, y_test=y_test,\
                         zlim=(-2,2), knots=[np.linspace(-1,1,2+i)]*2, ax=ax1, sz=10)
fig.suptitle(r"$\alpha=%.1E$"%(reg.alpha_))
fig.tight_layout()
fig.subplots_adjust(top=0.8)
plt.show()


fig=plt.figure(figsize=(16,4))
for i in range(4):
    ax1 = fig.add_subplot(1,4,1+i, projection="3d")
    reg = fit_plot3d_qsr(X_train = X_train, y_train = y_train, X_test=X_test, y_test=y_test,\
                         zlim=(-2,2), knots=[np.linspace(-1,1,2+i)]*2, ax=ax1, sz=10, alpha=1E5)
fig.suptitle(r"$\alpha=%.1E$"%(reg.alpha_))
fig.tight_layout()
fig.subplots_adjust(top=0.8)
plt.show()

fig=plt.figure(figsize=(16,4))
for i in range(4):
    ax1 = fig.add_subplot(1,4,1+i, projection="3d")
    reg = fit_plot3d_qsr(X_train = X_train, y_train = y_train, X_test=X_test, y_test=y_test,\
                         zlim=(-2,2), knots=[np.linspace(-1,1,2+i)]*2, ax=ax1, sz=10, alpha=1E-5)
fig.suptitle(r"$\alpha=%.1E$"%(reg.alpha_))
fig.tight_layout()
fig.subplots_adjust(top=0.8)
plt.show()
```
