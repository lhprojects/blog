# Math calculation

强度为
$$
A = \left|\sum \exp(i\phi_i)\right|^2 = A_s + A_c
$$

其中
$$
\phi_i = z_i k_z + y_i k_y + x_i k_x
$$

$$
A_s = \left(\sum_i p_i \sin\phi_i\right)^2
$$

$$
A_c = \left(\sum_i p_i \cos\phi_i\right)^2
$$

我们首先计算$A_s$,
$$
A_s = \sum_{ij} p_i p_j \sin\phi_i \sin\phi_j
$$

$\bar{p}_i$是一个随机变量，其数值为1时表示$i$处存在一个粒子，其数值为0时表示$i$处不存在一个粒子，$\bar{p}_i$是$i$处存在一个粒子的概率。
($\bar{p}_i$ is random variable. When its value is 1, it means there is a particle at position $i$; When its value is 0, it means there is no particle at position $i$; $\bar{p}_i$ is the probality of there is a particle at position $i$.)

然后$A_s$的期望为(Then the expectation of the $A_s$ is)
$$
\bar{A}_s = \sum_i (\bar{p_i^2} - \bar{p}^2_i) \sin^2\phi_i +  \sum_{ij} \bar{p}_i \bar{p}_j \sin\phi_i \sin\phi_j
$$


$$
\bar{A} = \sum_i (\bar{p_i^2} - \bar{p}^2_i) +\sum_{ij}  \bar{p}_i \bar{p}_j  (\sin\phi_i \sin\phi_j + \cos\phi_i \cos\phi_j)
$$
$$
\bar{A} = \sum_i \bar{p}_i(1 - \bar{p}_i) +  \sum_{ij}  \bar{p}_i \bar{p}_j (\sin\phi_i \sin\phi_j + \cos\phi_i \cos\phi_j)
$$


$$
\bar{p} = P(x,y,z) dx dy dz
$$ 
$$
\int dx dy dz P =  N
$$

$$
\bar{A} = N + \int\int dx'dy'dz' dxdydz P' P  (\sin\phi' \sin\phi + \cos\phi' \cos\phi)
$$

$$
\bar{A} = N + \int\int dx'dy'dz' dxdydz P' P \cos(\phi-\phi')
$$

$$
\bar{A} = N + \Re\int\int dx'dy'dz' dxdydz P' P \exp(ik_z(z-z') + i k_y(y-y') + i k_x(x-x'))
$$

$$
\bar{A} = N + \Re\int\int dx'dy'dz' dxdydz P' P \exp(ik_z z + i k_y y + i k_x x) \exp(-ik_z z - i k_y y - i k_x x)
$$

The (unitary, angular frequency) [fourier tranform](https://en.wikipedia.org/wiki/Fourier_transform) of $P$ is donated as $\overline{P}$
$$
\bar{A} = N + (2\pi)^{3/2} \Re \overline{P}(k_x, k_y, k_z) \overline{P}(-k_x, -k_y, -k_z)
$$

Note $N = (2\pi)^{3/2} \overline{P}(0,0,0)$

$$
\bar{A} = N \left(1 + N \frac{\Re \overline{P}(k_x, k_y, k_z) \overline{P}(-k_x, -k_y, -k_z)}{\overline{P}^2(0, 0, 0)}\right)
$$



## Form 2
为了分析干涉项对总体的影响，我们将上式写为


$$
\bar{A} = N \left(1 + (2\pi)^{3/2}\frac{\Re \overline{P}(k_x, k_y, k_z) \overline{P}(-k_x, -k_y, -k_z)}{\overline{P}(0, 0, 0)}\right)
$$

定义$I$

$$
\bar{A} = N \left(1 + (2\pi)^{3/2} \Re I\right)
$$

设密度$P$可以写为
$$
P(x) = c F(ax, ay, az)
$$

其中$c$的量纲为$L^{-1}$，从而$F$本身无量纲。
$a$的量纲为长度为$L^{-1}$，当$a\rightarrow 0 $时，物质体积趋于无穷大。

$$
\overline{P}(\omega) = c G(\omega/a, \omega/a, \omega/a)
$$

$$
I = \overline{P}^2(\omega)/\overline{P}(0) = cG^2(\omega/a, \omega/a, \omega/a)/G(0,0,0)
$$

我们预期$a\rightarrow 0$时候，对$\omega > 0，$我们有$I\rightarrow 0$。


## $k_x = 0$
当$k_x = 0$时，明显的，我们可以首先对$x$和$x'$积分。
可以将3维问题当作二维问题处理，并且二维的密度为$P(y,z)=\int dx P(x,y,z)$

$$
\bar{A} = N \left(1 + N \frac{\Re \overline{P}(k_y, k_z) \overline{P}(-k_y, -k_z)}{\overline{P}^2(0, 0)}\right)
$$


## Density distribution: Gauss
$$
F(x) = {\displaystyle e^{-\alpha x^{2}}\,}	
$$

$$
G(\omega) = {\displaystyle {\frac {1}{\sqrt {2\alpha }}}\cdot e^{-{\frac {\omega ^{2}}{4\alpha }}}}	
$$

$$
G(\omega)/G(0) = {\displaystyle e^{-{\frac {\omega ^{2}}{4\alpha }}}}	
$$

$$
I=  {\frac {c}{\sqrt {2\alpha }}} e^{-{\frac {\omega ^{2}}{2\alpha }}}
$$ 


## Density distribution: Box
$$
F(x) = {\displaystyle \operatorname {rect} (ax)\,}		
$$

$$
G(x) = {\displaystyle {\frac {1}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sinc} \left({\frac {\omega }{2\pi a}}\right)}	
$$

$$
G(\omega)/G(0) = {\displaystyle \operatorname {sinc} \left({\frac {\omega }{2\pi a}}\right)}	
= \frac{2\pi a}{\omega}\operatorname {sin} \left({\frac {\omega }{2\pi a}}\right)
$$

$$
I =  {\displaystyle {\frac {c}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sinc}^2 \left({\frac {\omega }{2\pi a}}\right)}	
= {\displaystyle \left(\frac {2\pi a}{\omega }\right)^2 {\frac {c}{\sqrt {2\pi a^{2}}}}\cdot \operatorname {sin}^2 \left({\frac {\omega }{2\pi a}}\right)}	
$$

## Density distribution: Exp

$$
F(x) = {\displaystyle \operatorname {e} ^{-a|x|}\,}	
$$
$$
G(\omega)/G(0) = \frac{a^2}{a^2 + \omega^2} 	
$$

## Density distribution: Circ

$$
F(x,y) = {\displaystyle \operatorname {circ} \left(a {\sqrt {x^{2}+y^{2}}}\right)}	
$$

$$
G(\omega_x, \omega_y)/G(0,0) = {\displaystyle {\frac {J_{1}\left({(1/a) \sqrt {\omega _{x}^{2}+\omega _{y}^{2}}}\right)} {(1/2)(1/a)\sqrt {\omega _{x}^{2}+\omega _{y}^{2}}}}}	
$$

$$
G(\omega_x, \omega_y)/G(0,0) = \left({(1/a)\sqrt {\omega _{x}^{2}+\omega _{y}^{2}}}\right)^{-3/2}
$$

# Numerical examination

```python
import numpy as np
import numba
import matplotlib.pyplot as plt

# light come form -z 
# scatter to y
@numba.njit
def cal_phi(z, y, sintheta, costheta, k):
    phi = (z * (1 - costheta) - y*sintheta)*k
    return phi

@numba.njit
def MC_A(thetas, N, lamb, width_z, width_y, model):
    k = 2*np.pi/lamb
    
    coss = np.full_like(thetas, 0.)
    sins = np.full_like(thetas, 0.)
    
    sinthetas = np.sin(thetas)
    costhetas = np.cos(thetas)
    for i in range(N):
        if model == "gauss":
            z = np.random.normal()*width_z
            y = np.random.normal()*width_y
        elif model == "box":
            z = (np.random.rand()-0.5)*width_z
            y = (np.random.rand()-0.5)*width_y
        
        for t, theta in enumerate(thetas): 
            phi = cal_phi(z, y, sinthetas[t], costhetas[t], k)
            #phi = 2*np.pi*np.random.rand()
            #print(z, y, phi)
            coss[t] += np.cos(phi)
            sins[t] += np.sin(phi)
            
    as_ = coss**2 + sins**2
            
    return as_/N

def cal_A(thetas, N, lamb, width_z, width_y, model):
    k = 2*np.pi/lamb
    
    As = np.full_like(thetas, 0.)
    
    ky = -np.sin(thetas)*k
    kz = (1-np.cos(thetas))*k
    if model == "gauss":
        def G_ratio(omega, width):
            alpha = 1/(2*width*width)
            return np.exp(-omega**2/(4*alpha))
    elif model == "box":
        def G_ratio(omega, width):
            a = 1/width
            x = omega/(2*np.pi*a)
            return np.sinc(x)
        
    return 1 + N*G_ratio(ky, width_y)*G_ratio(-ky, width_y)*G_ratio(kz, width_z)*G_ratio(-kz, width_z)
        
        
            

def plot(N, lamb, model="", 
         repeat=1, simulate=True,
         width_z=None, width_y=None, width=None,
         plot_range=1., plot_points=32, ax=None):
    thetas = np.linspace(-np.pi*plot_range, np.pi*plot_range,plot_points)    
    if width_z is None:
        width_z = width
    if width_y is None:
        width_y = width
    if width_z is None or width_y is None:
        assert False
        
    # simulate
    if simulate:
        Ass = np.empty(shape=(repeat, thetas.shape[0]))
        for r in range(repeat):
            Ass[r] = MC_A(thetas, N, lamb, width_z=width_z, width_y=width_y, model=model)
        As = Ass.mean(axis=0)
    
    Bs = cal_A(thetas, N, lamb, width_z=width_z, width_y=width_y, model=model)
    
    
    # begin plot
    show=False
    if ax is None:
        plt.figure(figsize=(4,4))
        ax = plt.gca()
        
    if simulate:
        ax.plot(thetas, As, label="numerical")
        
    ax.plot(thetas, Bs, label="theory")
    ax.set_yscale("log")
    ax.legend()
    
    if show:
        plt.show()
```

## Density distribution: Gauss

### change density

```python
figs, axes = plt.subplots(1,4, figsize=(16,4))
param = { "model": "gauss", "plot_points": 65, "plot_range": 0.1, "repeat": 1, "simulate": True, "lamb": 1}
plot(N=1000*1000, **param, width=   10, ax=axes[0])
plot(N=1000*1000, **param, width=  100, ax=axes[1])
plot(N=1000*1000, **param, width= 1000, ax=axes[2])
plot(N=1000*1000, **param, width=10000, ax=axes[3])
plt.show()
```
